```python
import numpy as np

# input variables
xs = [4,18, 26, 13, 0]
ys = [1, 0, 15, 16, 14]
alphas = [45, 120, 210, 260, 320]

# compute coeffs
ks = [1 / np.tan(a * np.pi / 180) for a in alphas]

# make the system of linear equations
A = np.array([[1, -k] for k in ks])
b = np.array([x - y*k for x, y, k in zip(xs, ys, ks)])

# solve using the method of least squares
solution = np.linalg.solve(A.T @ A, A.T @ b)
x_star, y_star = solution.round(3)
print("soluiton:", (x_star, y_star))

# find the error
error = np.sum(np.square(A @ solution - b))
print("error squared:", error.round(3))
```

```python
import matplotlib.pyplot as plt

plt.plot(x_star, y_star, '.', ms=20, color='black')

for bb, k in zip(b, ks):
    x = np.arange(-50, 50, 0.1)
    y = (x - bb) / k
    plt.plot(x, y)

plt.xlim(-20, 40)
plt.ylim(-20, 40)
plt.show()
```
